```python
import pandas as pd
```

```python
df = pd.read_csv('/content/ExtractedText14thApril.csv')
```

```python
corpus = []

for index, row in df.iterrows():
    corpus.append(row[4])
```

```python
print(corpus)
```

```python
tokenized_corpus = [doc.split(" ") for doc in corpus]
```

```python
print(tokenized_corpus)
```

```python
bm25 = BM25Okapi(tokenized_corpus)
```

```python
query = "Why does my adopted child have behavior challenges?"
tokenized_query = query.split(" ")
```

```python
doc_scores = bm25.get_scores(tokenized_query)
```

```python
doc_scores
```

```python
bm25.get_top_n(tokenized_query, corpus, n=5)
```

```python
# get the indices of the list in descending order based on their values
indices_descending = sorted(range(len(doc_scores)), key=lambda i: doc_scores[i], reverse=True)

# print the indices in descending order based on their values
print(indices_descending)
```

```python
TOP_N = 5

for i in range(TOP_N):
  print("Rank: " , i+1)
  print("Score: ", doc_scores[indices_descending[i]])
  print("Title: " , df.iloc[indices_descending[i] ,2])
  print("URL: " , df.iloc[indices_descending[i], 5])

  print("\n")
```

```python
import numpy as np

#print(np.argmax(doc_scores))
print("Title: " , df.iloc[np.argmax(doc_scores) ,2])
print("URL: " , df.iloc[np.argmax(doc_scores) ,5])
```
